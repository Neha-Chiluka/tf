```python
import pandas as pd
```

```python
url = "https://github.com/fenago/tf/blob/main/Chapter2-Loading_and_Processing_Data/dataset/Summary_of_Weather.csv"


df = pd.read_csv(url, low_memory=False)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
year_dummies = pd.get_dummies(df['Date'].dt.year, \
                                  prefix='year')
year_dummies
```

```python
month_dummies = pd.get_dummies(df['Date'].dt.month, \
                                   prefix='month')
month_dummies
```

```python
df = pd.concat([df, month_dummies, year_dummies], \
                   axis=1)
```

```python
df.drop(['Date', 'Precip','Snowfall', 'PoorWeather', 'PRCP',
        'TSHDSBRSGF', 'SNF'], axis=1, inplace=True)
```
