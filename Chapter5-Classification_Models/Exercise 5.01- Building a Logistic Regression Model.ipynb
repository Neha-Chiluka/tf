```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_url = "https://raw.githubusercontent.com/fenago/tf/main/Chapter5-Classification_Models/datasets/ion.csv"

data = pd.read_csv(data_url)
data.pop('Unnamed: 0')
data.head()
```

```python
data.info()
```

```python
reduce_map = {"Class": {"good":1, "bad":0} }
data.replace(reduce_map, inplace=True)
```

```python
data.head()
```

```python
y = data.pop('Class')
X = data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, 
                                                    random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
import tensorflow as tf
```

```python
tf.random.set_seed(8)
```

```python
model = tf.keras.Sequential()
```

```python
from tensorflow.keras.layers import Dense
```

```python
fc1 = Dense(512, input_shape=(34,), activation='relu')
```

```python
fc2 = Dense(512, activation='relu')
```

```python
fc3 = Dense(128, activation='relu')
```

```python
fc4 = Dense(128, activation='relu')
```

```python
fc5 = Dense(1, activation='sigmoid')
```

```python
model.add(fc1)
model.add(fc2)
model.add(fc3)
model.add(fc4)
model.add(fc5)
```

```python
model.summary()
```

```python
loss = tf.keras.losses.BinaryCrossentropy()
```

```python
optimizer = tf.keras.optimizers.Adam(0.001)
```

```python
model.compile(optimizer=optimizer, loss=loss)
```

```python
model.fit(X_train, y_train, epochs=5)
```

```python
preds = model.predict(X_test)
preds[:5]
```
